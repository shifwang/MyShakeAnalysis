```python
import time
from selenium import webdriver
import xlwt
```

```python
chrome = webdriver.Chrome() # start the chrome 
login_url = "http://www.nature.com/nature/journal/v512/n7515/full/nature12962.html" #nature paper link 1
login_url = "http://www.nature.com/nature/journal/v512/n7515/full/nature13424.html" #nature paper link 2
login_url = "http://www.nature.com/articles/ncomms6897" #nature paper link 3

chrome.get(login_url) #open the website
```

```python
tmp = chrome.find_elements_by_class_name("collapsed")
if len(tmp) > 0:
    tmp[0].click(); # click the et al. to see the whole list
```

```python
maxiter = 500
book = xlwt.Workbook(encoding="utf-8")
sheet1 = book.add_sheet("Sheet 1")
#article_heading = chrome.find_element_by_class_name("article-heading").text
#sheet1.write(0, 0, article_heading)
for i in range(1, maxiter):
    print(i)
    tmp = chrome.find_elements_by_css_selector("a[href='#auth-%d']"%i)
    if len(tmp) > 0:
        name = tmp[0].find_element_by_css_selector("span").text
        print(name)
        sheet1.write(i, 0, name)
        tmp[0].click()
        time.sleep(0.5)
        # following could be popup-auth or author-auth
        #affl = chrome.find_element_by_id('author-auth-%d'%i).find_element_by_class_name("popup-content").find_element_by_class_name("author-affiliations").find_element_by_css_selector('li').text
        affl = chrome.find_element_by_id('popup-auth-%d'%i).find_element_by_css_selector('section ul li').text
        sheet1.write(i, 1, affl)
    else:
        break
book.save("trial.xls")
```
